```python
import os

# Download TREC Covid Dataset
dataset = pt.datasets.get_dataset('irds:cord19/trec-covid')
pt_index_path = './indices/cord19'

# Create index for TREC Covid Dataset
indexer = pt.index.IterDictIndexer(pt_index_path, blocks=True)
index_ref = indexer.index(dataset.get_corpus_iter(), 
                          fields=['title', 'abstract'],
                          meta=('docno','doi','title'))
  
index = pt.IndexFactory.of(index_ref)
```

```python
# Create baseline based on BM25 and Top 1000 results
bm25 = pt.BatchRetrieve(index_ref , wmodel='BM25', num_results=1000, metadata=["docno","doi","title"])
res = bm25.transform(dataset.get_topics('title'))
res
```

```python
# Extract TREC Covid columns from dataset into list

list_for_df= []
for j in dataset.get_corpus_iter():
  list_for_df.append(j)
  #print(j)
```

```python
import pandas as pd
```

```python
# Create dataframe from the TREC Covid Dataset

df = pd.DataFrame(list_for_df)
```

```python
# Filter dataframe based on documents that appear in the Top 1000 from the baseline

df = df[df["docno"].isin(res["docno"].tolist())]
```

```python
# Filter documents that have an empty string value for the DOI column

df =  df[df["doi"] != ""]
```

```python
# Remove duplicated documents from the dataframe

df = df.drop_duplicates("docno", keep="first")
```

```python
# Install third party python library to access Semantic Scholar API
pip install semanticscholar
```

```python
from semanticscholar import SemanticScholar
```

```python
# Enter API Key here, to increase the amount of scrapable documents
s2_api_key = ''

sch = SemanticScholar(api_key=s2_api_key)

# Creating list, where metadata for every document are saved
metadata_columns = []

# Specify the metadata that should be retrieved from the API
fields_to_scrape = ['citations.authors',"authors",'authors.affiliations', 'fieldsOfStudy', 'journal','authors.name','s2FieldsOfStudy']

# Iterate through dataframe with the documents from the 50 result lists 
for i,k in df.iterrows():
  # Access the API with the DOI of the paper
  paper = sch.get_paper(paper_id = k["doi"], fields=fields_to_scrape)

  # Create temporary lists for the various retrieved metadata
  authors = []
  ref_authors = []
  field_of_study= []
  journal = []
  affiliations = []
  s2FieldsOfStudy = []
   
  # Add metadata to the corresponding list, if metadata is provided by the API 
  try:
    for j in paper["fieldsOfStudy"]:
      field_of_study.append(j)
  except:
    field_of_study.append("NaN")
  try:
    for j in paper["s2FieldsOfStudy"]:
      s2FieldsOfStudy.append(j)
  except:
    s2FieldsOfStudy.append("NaN")
 
  try:
    journal.append(paper["journal"]["name"])
  except:
    journal.append("NaN")

  try:
    for j in paper["authors"]:
      authors.append(j)
  except:
    authors.append("NaN")
  try:
    for j in paper["citations"]:
      ref_authors.append(j) 
  except:
    ref_authors.append("NaN")
  
  # Add the temporary lists to the metadata_columns list
  metadata_columns.append([k["docno"],authors,ref_authors,s2FieldsOfStudy, field_of_study,journal ])
 
```

```python
import pandas as pd
```

```python
# Create dataframe with the additional metadata from list

df = pd.DataFrame(metadata_columns)
```
